### **Dificulty

The difficulty level is based on the completion statistics of Advent of Code (https://adventofcode.com/2022/stats) as of April 1st 2023. At that moment 12'042 users solved all puzzles.

The difficulty levels are grouped by the following ranges:

```python
from utils.stats import DifficultyLevelBinner

gold_stars = [
    250841, 211106, 182985, 169638, 145888, 144651, 105547, 100445, 77450,
    79867, 64803, 55577, 48433, 46157, 37311, 21568, 20074, 23858, 16157,
    20724, 21584, 14245, 17233, 15079, 12042
]
silver_stars = [
    11187, 10008, 9400, 3090, 2646, 1375, 2267, 6837, 10546, 4945, 8557,
    944, 1109, 888, 5421, 5272, 5933, 4670, 884, 956, 2876, 4872, 237,
    205, 6520
]

binner = DifficultyLevelBinner(
    gold_stars, silver_stars,
    difficulty_levels=["Very Easy", "Easy", "Normal", "Hard"],
    difficulty_weights=[1, 1, 1, 1],
    prediction_interval=90  # consider 10% of the values as outliers
)

for (
    difficulty_level, (min_range_value, max_range_value)
) in binner.difficulty_level_ranges.items():
    print(f"{difficulty_level}: [{min_range_value}, {max_range_value}]")
```

Hereby, each range defines the minimum and maximum number of users who completed a challenge.

```python
import pandas as pd

print("Difficulty Levels - Part 1")

day_stats = []
for day_index, completion_count in enumerate(
    binner.part1_completion_counts
):
    difficulty_level = binner.get_difficulty_level(completion_count)
    day_stats.append(
        [f"Day {day_index + 1}", difficulty_level, completion_count]
    )

pd.DataFrame(
    day_stats, columns=["Day", "Difficulty Level", "Completed By # Users"]
)
```

```python
print("Difficulty Levels - Part 2")

day_stats = []
for day_index, completion_count in enumerate(
    binner.part2_completion_counts
):
    difficulty_level = binner.get_difficulty_level(completion_count)
    day_stats.append(
        [f"Day {day_index + 1}", difficulty_level, completion_count]
    )

pd.DataFrame(
    day_stats, columns=["Day", "Difficulty Level", "Completed By # Users"]
)
```
